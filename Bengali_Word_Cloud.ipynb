```python
import pandas as pd

def separate_and_save_dataframes(df, base_filename='data'):
    """
    Separates a DataFrame into three DataFrames based on the 'label' column and saves them as CSV files.

    Parameters:
        df (pd.DataFrame): The input DataFrame with columns 'text' and 'label'.
        base_filename (str): The base name for the output files (default is 'data').

    Returns:
        tuple: Three DataFrames corresponding to 'positive', 'negative', and 'neutral' labels.
    """
    # Separate data into three different DataFrames based on the label and drop the 'label' column
    positive_df = df[df['label'] == 'positive'].drop(columns=['label'])
    negative_df = df[df['label'] == 'negative'].drop(columns=['label'])
    neutral_df = df[df['label'] == 'neutral'].drop(columns=['label'])

    # Save each DataFrame to a CSV file
    positive_df.to_csv(f'{base_filename}_positive.csv', index=False)
    negative_df.to_csv(f'{base_filename}_negative.csv', index=False)
    neutral_df.to_csv(f'{base_filename}_neutral.csv', index=False)

    return positive_df, negative_df, neutral_df

# Example usage
# df = pd.read_csv('your_input_file.csv')  # Replace 'your_input_file.csv' with the actual file path
# positive_df, negative_df, neutral_df = separate_and_save_dataframes(df, base_filename='output_data')
```

```python
import pandas as pd
from collections import Counter
import re

def find_common_words_for_each(positive_df, negative_df, neutral_df):
    """
    Finds and counts common words separately in the 'text' columns of three DataFrames.

    Parameters:
        positive_df (pd.DataFrame): DataFrame containing positive text data.
        negative_df (pd.DataFrame): DataFrame containing negative text data.
        neutral_df (pd.DataFrame): DataFrame containing neutral text data.

    Returns:
        tuple: Three DataFrames containing word counts for each category (positive, negative, neutral).
    """
    def count_words(df):
        # Tokenize the text into individual words and count occurrences
        word_counts = Counter()
        for text in df['text']:
            # Use regex to find words and convert to lowercase
            words = re.findall(r'\b\w+\b', text.lower())
            word_counts.update(words)
        # Convert the Counter to a DataFrame
        return pd.DataFrame(word_counts.items(), columns=['word', 'count']).sort_values(by='count', ascending=False).reset_index(drop=True)

    # Find common words for each DataFrame
    positive_common_words = count_words(positive_df)
    negative_common_words = count_words(negative_df)
    neutral_common_words = count_words(neutral_df)

    return positive_common_words, negative_common_words, neutral_common_words

# Example usage
# Assuming you already have the DataFrames positive_df, negative_df, neutral_df
positive_common_words, negative_common_words, neutral_common_words = find_common_words_for_each(positive_df, negative_df, neutral_df)

#Display the results
print("Positive Common Words:\n", positive_common_words)
print("\nNegative Common Words:\n", negative_common_words)
print("\nNeutral Common Words:\n", neutral_common_words)
```

```python
import pandas as pd
from collections import Counter
import re

def find_common_words_for_each(positive_df, negative_df, neutral_df):
    """
    Finds and counts common words separately in the 'text' columns of three DataFrames.

    Parameters:
        positive_df (pd.DataFrame): DataFrame containing positive text data.
        negative_df (pd.DataFrame): DataFrame containing negative text data.
        neutral_df (pd.DataFrame): DataFrame containing neutral text data.

    Returns:
        tuple: Three DataFrames containing word counts for each category (positive, negative, neutral).
    """

    def count_words(df):
        """Counts the occurrences of each word in the 'text' column of the DataFrame."""
        word_counts = Counter()  # Initialize a Counter to count words
        for text in df['text']:
            # Tokenize the text into individual words using regex, converting to lowercase
            words = re.findall(r'\b\w+\b', text.lower())
            word_counts.update(words)  # Update the Counter with the words
        # Convert the Counter to a DataFrame and sort by count in descending order
        return pd.DataFrame(word_counts.items(), columns=['word', 'count']).sort_values(by='count', ascending=False).reset_index(drop=True)

    # Find common words for each DataFrame
    positive_common_words = count_words(positive_df)
    negative_common_words = count_words(negative_df)
    neutral_common_words = count_words(neutral_df)

    # Save the common words DataFrames to CSV files
    positive_common_words.to_csv('positive_common_words.csv', index=False)
    negative_common_words.to_csv('negative_common_words.csv', index=False)
    neutral_common_words.to_csv('neutral_common_words.csv', index=False)

    return positive_common_words, negative_common_words, neutral_common_words

# Example usage
# Assuming you already have the DataFrames positive_df, negative_df, neutral_df
positive_common_words, negative_common_words, neutral_common_words = find_common_words_for_each(positive_df, negative_df, neutral_df)

# Print the common words if needed
print("Positive Common Words:\n", positive_common_words)
print("\nNegative Common Words:\n", negative_common_words)
print("\nNeutral Common Words:\n", neutral_common_words)
```
